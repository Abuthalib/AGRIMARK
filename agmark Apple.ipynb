```python
# removing outliers

threshold = 1.5
df_no_outliers = num_col.copy()
for col in num_col:
    Q1 = num_col[col].quantile(0.25)
    Q3 = num_col[col].quantile(0.75)
    IQR = Q3 - Q1
    outlier_mask = (num_col[col] < (Q1 - threshold * IQR)) | (num_col[col] > (Q3 + threshold * IQR))
    df_no_outlier = df_no_outliers[~outlier_mask]

for cols in df_no_outlier.columns:
    plt.figure(figsize=(10,6))
    sns.boxplot(data = df[cols])
    plt.title(f"Box Plot for {cols}")
    plt.show()
```

```python
scaler = StandardScaler()
scaler = scaler.fit(df_for_training)
scaled_data = scaler.transform(df_for_training)
```

```python
scaled_data
```

```python
time_steps = 7
```

```python
X,y =[],[]
for i in range(len(scaled_data) - time_steps):
    X.append(scaled_data[i:i+time_steps])
    y.append(scaled_data[i+time_steps][-1])
```

```python
X = np.array(X)
y = np.array(y)
```

```python
# Split the dataset into training and testing sets
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

X_train.shape, X_test.shape , y_train.shape , y_test.shape
```

```python
n_timesteps = 7
n_features = 4
```

```python
# Define the LSTM model
model = Sequential()

# Add an LSTM layer with 64 units and return sequences
model.add(LSTM(64, return_sequences=True, input_shape=(n_timesteps, n_features)))

# Add a second LSTM layer with 32 units
model.add(LSTM(32))

# Add a new dense layer with 64 units and relu activation
model.add(Dense(64, activation='relu'))

# Add the output layer with a single unit and linear activation (for regression)
model.add(Dense(1, activation='linear'))

# Compile the model
model.compile(optimizer='adam', loss='mse')

# Print the model summary
print(model.summary())
```

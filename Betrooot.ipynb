```python
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler
import seaborn as sns
```

```python
date_col = df.select_dtypes(include=['datetime'])
date_col
```

```python
df_for_training = num_col
df_for_training
```

```python
df_for_plot = df_for_training.tail(5000)
```

```python
df_for_plot.plot.line()
```

```python
scaler = StandardScaler()
scaler = scaler.fit(df_for_training)
scaled_data = scaler.transform(df_for_training)
```

```python
scaled_data
```

```python
time_steps = 7
```

```python
X,y =[],[]
for i in range(len(scaled_data) - time_steps):
    X.append(scaled_data[i:i+time_steps])
    y.append(scaled_data[i+time_steps][-1])
```

```python
X = np.array(X)
y = np.array(y)
```

```python
# Split the dataset into training and testing sets
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

X_train.shape, X_test.shape , y_train.shape , y_test.shape
```

```python
n_timesteps = 7
n_features = 4
```

```python
# Define the LSTM model
model = Sequential()

# Add an LSTM layer with 64 units and return sequences
model.add(LSTM(64, return_sequences=True, input_shape=(n_timesteps, n_features)))

# Add a second LSTM layer with 32 units
model.add(LSTM(32))

# Add a new dense layer with 64 units and relu activation
model.add(Dense(64, activation='relu'))

# Add the output layer with a single unit and linear activation (for regression)
model.add(Dense(1, activation='linear'))

# Compile the model
model.compile(optimizer='adam', loss='mse')

# Print the model summary
print(model.summary())
```

```python

# Train the model
model.fit(X_train, y_train, epochs=20, batch_size=32, verbose=1)
```

```python
# Evaluate the model
loss = model.evaluate(X_test, y_test)
print('Testing Loss:', loss)
```

```python
# Assuming the model is already trained

# Perform inference on the testing data
predictions = model.predict(X_test)

# You may want to inverse transform the predicted values to get them back to the original scale
predicted_prices = scaler.inverse_transform(np.concatenate((X_test[:, -1, :-1], predictions.reshape(-1, 1)), axis=1))[:, -1]

# Print the predicted prices
print("Predicted Prices:")
print(predicted_prices)

# Compare the predicted prices with the actual prices
actual_prices = scaler.inverse_transform(np.concatenate((X_test[:, -1, :-1], y_test.reshape(-1, 1)), axis=1))[:, -1]
print("Actual Prices:")
print(actual_prices)
```

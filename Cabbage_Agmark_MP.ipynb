```python

for cols in num_col.columns:
    plt.figure(figsize=(10,6))
    sns.boxplot(data = df[cols])
    plt.title(f"Box Plot for {cols}")
    plt.show()
```

```python
num_col.shape
```

```python
# removing outliers

threshold = 1.5
df_no_outliers = num_col.copy()
for col in num_col:
    Q1 = num_col[col].quantile(0.25)
    Q3 = num_col[col].quantile(0.75)
    IQR = Q3 - Q1
    outlier_mask = (num_col[col] < (Q1 - threshold * IQR)) | (num_col[col] > (Q3 + threshold * IQR))
    df_no_outlier = df_no_outliers[~outlier_mask]
```

```python
for cols in df_no_outlier.columns:
    plt.figure(figsize=(10,6))
    sns.boxplot(data = df[cols])
    plt.title(f"Box Plot for {cols}")
    plt.show()
```

```python
df_no_outlier.shape
```

```python
# Histplot
for col in num_col.columns:
    plt.figure(figsize=(10,6))
    sns.histplot(data=num_col[col],kde=True)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
num_corr = num_col.corr()
num_corr
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(num_corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

# multi variable analysis

```python
#price per districts
max_prices_district = df.groupby('District Name')['Max Price (Rs./Quintal)'].max().sort_values(ascending=False)


plt.figure(figsize=(10, 6))
max_prices_district.plot(kind='bar')
plt.title('Districts with Maximum Prices')
plt.xlabel('District Name')
plt.ylabel('Max Price (Rs./Quintal)')
plt.xticks(rotation=90)
plt.show()
```

```python
num_col.shape
```

```python
# removing outliers

threshold = 1.5
df_no_outliers = num_col.copy()
for col in num_col:
    Q1 = num_col[col].quantile(0.25)
    Q3 = num_col[col].quantile(0.75)
    IQR = Q3 - Q1
    outlier_mask = (num_col[col] < (Q1 - threshold * IQR)) | (num_col[col] > (Q3 + threshold * IQR))
    df_no_outlier = df_no_outliers[~outlier_mask]
```

```python
for cols in df_no_outlier.columns:
    plt.figure(figsize=(10,6))
    sns.boxplot(data = df[cols])
    plt.title(f"Box Plot for {cols}")
    plt.show()
```

```python
# Lowest and Higest modal price based on market
highest_modal_price = df.groupby('Market Name')['Modal Price (Rs./Quintal)'].max()
lowest_modal_price = df.groupby('Market Name')['Modal Price (Rs./Quintal)'].min()
for market, highest_price in highest_modal_price.items():
    lowest_price = lowest_modal_price[market]
    print(f"Market: {market}")
    print(f"Highest Modal Price: {highest_price}")
    print(f"Lowest Modal Price: {lowest_price}")
    print("*" * 50)
```

```python
# Lowest and Higest modal price based on Districts
highest_modal_price = df.groupby('District Name')['Modal Price (Rs./Quintal)'].max()
lowest_modal_price = df.groupby('District Name')['Modal Price (Rs./Quintal)'].min()
for dist, highest_price in highest_modal_price.items():
    lowest_price = lowest_modal_price[dist]
    print(f"District: {dist}")
    print(f"Highest Modal Price: {highest_price}")
    print(f"Lowest Modal Price: {lowest_price}")
    print("*" * 50)
```

```python
# Plot top 10 markets with higest modal prices.

top_10_highest_prices = df.groupby('Market Name')['Modal Price (Rs./Quintal)'].max().nlargest(10)

plt.figure(figsize=(10, 6))
top_10_highest_prices.plot(kind='bar', color='skyblue')
plt.title('Top 10 Markets with Highest Modal Prices')
plt.xlabel('Market Name')
plt.ylabel('Modal Price (Rs./Quintal)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
